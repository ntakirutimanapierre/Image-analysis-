# Before doing image derivatives, let's see how to import a folder of images.

There are three steps to follow in importing a folder of images.
1. Accest the image folder by using folder/ path---os module
2. Read the images using --opencv package imread()
3. Display the images using imshow() matplotlib module

```python
#import the necessary packages
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

#a function that will always help us load the images
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images

#To load the images of flowers
folder='flowers'
listt=load_images_from_folder(folder)
```

```python
for j in range(4,10):
    picture=listt[j]
    print(picture.shape)
    plt.imshow(picture)
    plt.show()
```
